#### What is Julia and why should I care? What is Julia good at?

* An exciting new language made specifically for scientific computing 
* Interpreted (like Matlab/Python), but also (JIT-)compiled

* Easy to learn, write, and maintain, but designed with performance in mind (2-3x C speed)
    * Define functions wherever you want
    * Functions with optional, keyword, defualt arugments
    * Organize code in modules/packages

* Core concept is multiple dispatch -- specialize function behavior based on types of *all* arguments (more later)
* Parallel programming features built into the language - easy and inexpensive scalability (more later)

* Most of base Julia is written in Julia itself
    * User code is set on "same level" as language and library code
    * With smart algorithms, will have same performance as the language itself
    * If you know Julia, you can look at source and see how the language itself is implemented

## Multiple Dispatch

* Julia has an advanced expressive type system
* ... but you can only think about it when you want
* Functions are applied by matching function name and types of arguments
* The most specific match possible is called
* Consider a contrived example that demonstrates the concept

```julia
f(a::Any, b) = "fallback"
f(a::Number, b::Number) = "a and b are both numbers"
f(a::Number, b) = "a is a number"
f(a, b::Number) = "b is a number"
f(a::Integer, b::Integer) = "a and b are both integers"
```

```julia
f(randn(4, 4), "foo")
```

```julia
f(1, 1.0)
```

```julia
f(1.0, "foo")
```

```julia
f("foo", 1)
```

```julia
f(1, 2)
```

## Parallel Computing

* Building blocks for parallel computaiton are built in to the language
    - Ability to programmatically start additional local or remote processes (`addprocs`)
    - Operate on distributed arrays (`DArray`)
    - Shared memory arrays for local processes (`SharedArray `)
    - Parallel `for` loops and `map`
* Easy and free to install in the cloud (`AWS.jl`, `GCloud.jl`)

### `@parallel for`

* Suppose we wanted to simulate the number of heads in `n` coinflips
* We might write the following function:

```julia
function count_heads(n)
    c::Int = 0
    for i=1:n
        c += randbool()
    end
    c
end

count_heads(10)  # warm up compiler
@timeit count_heads(50000000)  # run timing
```

* Now consider the equivalent parallel version of this function:

```julia
function count_heads_par(n)
    @parallel (+) for i=1:n
        int(randbool())
    end
end

addprocs(5)  # add five local processes
count_heads_par(10)  # warm up compiler
@timeit count_heads_par(50000000)  # run timing
```

Things to notice:

* The parallel version was shorter than the original
* Ran more than 4 times faster

### Case study: solving models in QuantEcon.jl

* The QuantEcon.jl library has a number of classic Macroeconomic models
* Most have these common functions defined:
    - `bellman_operator`: perform one iteration of natural operator defined by the Bellman equation
    - `init_values`: produce initial values for discretized value function
* These building blocks suggest a common (if naive) algorithm for solving a model
    - Compute initial guess for value function from `init_values`
    - Iterate on `bellman_operator` until you reach a fixed point
* In code this looks like

```julia
function solve_vf(m::AbstractModel)
    init = init_values(m)
    f(x) = bellman_operator(m, x)
    return compute_fixed_point(f, init)
end 
```

* *If* we happen to know a better routine for solving a specific model, we can implement it as a special case

```julia
@everywhere using QuantEcon, QuantEcon.Models

c = CareerWorkerProblem() # (Neal, 1999)
ifp = ConsumerProblem()   # income fluctuation problem -- many refs
jv = JvWorker()           # (Jovanovic, 1979)
odu = SearchProblem()     # (Ljungqvist and Sargent, 2012)
gm = GrowthModel()        # standard growth model -- many refs

models = [c, ifp, jv, odu, gm];
```

```julia
for m in models
    println(@which solve_vf(m))
end
```

```julia
# Run through one iteration of each to warm up compiler
map(m -> solve_vf(m, verbose=false, err_tol=Inf), models);

# solve in serial
@time map(m -> solve_vf(m, verbose=false, max_iter=1000), models);
```

```julia
# do the same in parallel
pmap(m -> solve_vf(m, verbose=false, err_tol=Inf), models);
@time pmap(m -> solve_vf(m, verbose=false, max_iter=1000), models);
```

* Not quite 5 times faster... why?
* Some models take longer to solve than others. So some processes finished long before 10 seconds, but had to wait for others

* Suppose we wanted to solve the Ljungqvist and Sargent search with unknown offer distribution model for agents of various discount factors...
* Easy!

```julia
odu_mods = [SearchProblem(bet=i) for i in linspace(0.9, 0.99, 500)];
@time results = pmap(m -> solve_vf(m, verbose=false, max_iter=1000), odu_mods);
```

```julia
using PyPlot

function plot_results(results, odu_mods, n=5)
    n_res = size(results, 1)
    inds = int(linspace(1, n_res, n))
    fig, ax = subplots()
    for i in inds
        mod = odu_mods[i]
        bet = round(mod.bet, 2)
        pi_grid = mod.pi_grid
        ax[:plot](pi_grid, results[i], label=L"$\beta=$"*"$bet")
    end
    ax[:set_xlabel](L"Probabililty of 'bad' model ($\pi$)")
    ax[:set_ylabel](L"Reservation wage ($\bar{w}$)")
    ax[:legend]()
    nothing
end

plot_results(results, odu_mods);
```
