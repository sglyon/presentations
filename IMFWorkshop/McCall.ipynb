## A simple problem: McCall Search model


Now that we know what types are in Julia, let's create one to describe the McCall search model and hook it in to the `QuantEcon.jl` framework.

Knowing that is our goal, we need to import some tools from `QuantEcon.jl`:

````julia
using QuantEcon, QuantEcon.Models
import QuantEcon.Models: bellman_operator, init_values
````

### The Model

I will give a *very breif* review of the McCall search model so we are all on the same page.

An unemployed worker searches for a job each period and receives an offer to work forever at a wage $w$, which has some distribution $F(w) = \text{prob} \{w < W \}$, with $F(0)= 0, F(B) = 1$ for $B < \infty$

Each period the worker is unemployed he receives unemployment benefits equal to $c$.

The worker seeks to maximize $\sum_{t=0}^{\inf} \beta^t y_t$, where $\beta \in (0, 1)$ is a discount factor and $y_t$ is the workers income in period $t$.

Let $v(w)$ be the value of our unemployed worker who is deciding whether to accept or reject an offer at wage $w$. We write $v(w)$ as

$$
v(w) = \max_{\text{accept, reject}} \left\{\frac{w}{1 - \beta}, \beta \int_0^{B} v (w'), dF(w')\right\}
$$


### Implementation

To implement this model, we will construct a type `McCallModel` as follows:

````julia
immutable McCallModel <: AbstractModel
    w::Vector        # vector of wages
    F::Vector        # probability of each w
    β::Real          # discount factor
    c::Real          # reservation wage
    v_accept::Vector # value of accepting at each wage

    # Constructor to fill in v_accpet.
    function McCallModel(w::Vector, F::Vector, β::Real, c::Real)
        new(w, F, β, c, w./(1 - β))
    end
end

# constructor with keyword arguments and default values
function McCallModel(;w::Vector=linspace(0.1, 50.0, 5000),     # evenly spaced
                      F::Vector=fill(1/size(w,1), size(w,1)),  # uniform
                      β::Real=0.95,
                      c::Real=10.0)
    return McCallModel(w, F, β, c)
end
````

A few comments about our `McCallModel` type are in order:

* We used the keyword `immutable` instead of `type` to tell the compiler we don't expect the *fields* (`w`, `F`, `β`,...) to change after the object is created. We did this because there is a special relationship between the `w`, `β` and `v_accept` fields
* We are defining `McCallModel` to be a subtype of the `AbstractModel` type defined in QuantEcon.jl. We will see why shortly...
* We have defined two *constructors* (functions that create `McCallModel` objects) for our type:
    * An *outer-constructor* that appears after the `end` corresponding to the type declaration. This is standard practice in Julia.
    * An *inner-constructor* that appears on the inside of our type declaration. 
        * This is less common, but is a techinique used to make sure that certain properties of the object hold. In this case we wanted to make sure that `v_accept` is defined properly as `w / (1 - β)`.
        * Also notice the use of keyword arugments and default values here. Keyword arguments follow a semicolon (`;`) in the function arugment list and enable us to change specific arguments by name, leaving the others unchanged.
        
Now, to actually solve this model, we will call upon some machinery from QuantEcon.jl. In this library there are a number of models that have each implemented functions called `bellman_operator` and `init_values`. 

The `bellman_operator` function takes the model object and a discretized representation of the current value function and returns the next iteration of the value function. 

`init_values` takes the model object and produces an initial discritization of the value function that can be used to start off the iterations.

Using these two functions, QuantEcon.jl is able to solve generic models with an algorithm that is equivalent to the code below (the actual implementation comes with a few extra bells and whistles, but follows the same algorithm):

```julia
function solve_vf(m::AbstractModel)
    init = init_values(m)
    f(x) = bellman_operator(m, x)
    return compute_fixed_point(f, init)
end  
```

A few comments:

* Notice that the argument of the function is of type `AbstractModel`. This means that any type that is declared with  `type <: AbstractModel` or `immutable <: AbstractModel` will have a method `solve_vf` defined for it. 
* In order for this function to run without errors, the subtype of `AbstractModel` must implement both `init_values` and `bellman_operator`. 

Recall that we defined `immutable McCallModel <: AbstractModel`. This means that if we just define `bellman_operator(m::McCallModel, v)` and `init_values(m::McCallModel)`, we can let QuantEcon.jl solve this model for us. Let's give it a try!

````julia
function bellman_operator(m::McCallModel, v0)
    v_reject = m.c + m.β * sum(v0 .* m.F)
    max(m.v_accept, v_reject)  # return keyword optional
end

# start at value of accepting mean wage
init_values(m::McCallModel) = fill(mean(m.w)/(1 - m.β), size(m.w, 1))
````

At this point we are ready to construt an object of our `McCallModel` type and compute the solution!

````julia
m = McCallModel()  # use default values
vf = solve_vf(m);
````
