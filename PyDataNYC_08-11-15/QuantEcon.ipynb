#### Example: Auto-regressive moving-average models

- Models parameterized by coefficients $\{\phi\}$ and $\{\theta\}$ and $\sigma$: 

$$X_t = \sum_{p=1}^{N} \phi_p X_{t-p} + \epsilon_t  + \sum_{q=1}^M \theta_q \epsilon_{t-q}$$

- Where $\epsilon_t \stackrel{iid}{\sim} N(0, \sigma^2) \; \forall t$
- Represented in `QuantEcon` by the `ARMA` class

```python
from quantecon import ARMA
%matplotlib inline
```

```python
ARMA?
```

- The `ARMA` class provides methods to visualize a model's:
    - Impulse response function
    - Simulated Time series
    - Autocovariance function
    - Spectral density


- Example: $X_t = 0.5 X_{t-1} + \epsilon_t - 0.8 \epsilon_{t-2}$

```python
m = ARMA(phi=[0.5], theta=[-0.8])
m.quad_plot() 
```

- Let's leverage the fact that we are in a notebook and make it interactive

```python
from IPython.html.widgets import interact

def plot_arma11_model(phi_1, theta_1):
    m = ARMA(phi=[phi_1], theta=[theta_1])
    m.quad_plot() 
```

```python
interact(plot_arma11_model, phi_1=(-0.99, 0.99, 0.1), theta_1=(-5.0, 5.0, 0.1))
```
